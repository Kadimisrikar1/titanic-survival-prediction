```python
#Import libraries

import pandas as pd 
from sklearn.impute import SimpleImputer
from sklearn import tree
from sklearn import model_selection
```

```python
#loading the datasets

titanic_train = pd.read_csv("train.csv")
titanic_train.shape
titanic_train.info()
```

```python
titanic_train.head(22)
```

```python
titanic_test = pd.read_csv('test.csv')
titanic_test.shape
titanic_test.info()
```

```python
titanic_test.head(22)
```

```python
#there is no survived attribute in the test dataset
titanic_test.Survived = None
```

```python
#concatinating the train and test datasets
titanic = pd.concat([titanic_train, titanic_test])
titanic.shape
titanic.info()
```

```python
titanic.head(22)
```

```python
#spliting the format of the title

def extract_title(name):
    return name.split(',')[1].split('.')[0].strip()
```

```python
titanic['Title'] = titanic['Name'].map(extract_title)
```

```python
#Imputing the missing values by using the imputer
mean_imputer = SimpleImputer() 
mean_imputer.fit(titanic_train[['Age','Fare']])
```

```python
titanic.head(10)
```

```python
titanic[['Age','Fare']] = mean_imputer.transform(titanic[['Age','Fare']])
```

```python
#For age variable converting the numerical column into categorical column 

def convert_age(age):
    if(age >= 0 and age <= 10): 
        return 'Child'
    elif(age <= 25): 
        return 'Young'
    elif(age <= 50): 
        return 'Middle'
    else: 
        return 'Old'
```

```python
titanic['Age_Cat'] = titanic['Age'].map(convert_age)
```

```python
titanic['FamilySize'] = titanic['SibSp'] +  titanic['Parch'] + 1   #for family size to be not equal to zero adding 1
```

```python
#for family size converting the numerical column into categorical column 

def convert_familysize(size):
    if(size == 1): 
        return 'Single'
    elif(size <=3): 
        return 'Small'
    elif(size <= 6): 
        return 'Medium'
    else: 
        return 'Large'
```

```python
titanic['FamilySize_Cat'] = titanic['FamilySize'].map(convert_familysize)
```

```python
#one-hot encoding
#in the process of one-hot encoding creating the dummy variables for the following columns 
titanic1 = pd.get_dummies(titanic, columns=['Sex','Pclass','Embarked', 'Age_Cat', 'Title', 'FamilySize_Cat'])
titanic1.shape
titanic1.info()
```

```python
#creating a dataframe by dropping some columns
titanic2 = titanic1.drop(['PassengerId','Name','Age','Ticket','Cabin','Survived'], axis=1, inplace=False)
titanic2.shape 
titanic2.info()
```

```python
X_train = titanic2[0:891] #0 t0 891 records
X_train.shape
X_train.info()
y_train = titanic_train['Survived']
```

```python
tree_estimator = tree.DecisionTreeClassifier()
```

```python
#values of the dt parameters
dt_grid = {'max_depth':[15, 16, 17], 'min_samples_split':[2,3], 'criterion':['gini','entropy']}
#dt_grid = {'max_depth':list(range(10,30)), 'min_samples_split':list(range(2,8)), 'criterion':['gini','entropy']}
```

```python
param_grid = model_selection.GridSearchCV(tree_estimator, dt_grid, cv=15) #Evolution of tee
param_grid.fit(X_train, y_train) #Building the tree
param_grid.cv_results_
print(param_grid.best_score_) #Best score
print(param_grid.best_params_)
```

```python
fi_df = pd.DataFrame({'feature':X_train.columns, 'importance':  param_grid.best_estimator_.feature_importances_}) #You may notice that feature	importance "Title_Mr" has more importance
print(fi_df)
```

```python
X_test = titanic2[titanic_train.shape[0]:] #shape[0]: means 0 index to n index. Not specifying end index is nothing but till nth index
X_test.shape
X_test.info()
titanic_test['Survived'] = param_grid.predict(X_test)
```

```python
titanic_test.to_csv('paramimp.csv',columns=['PassengerId','Survived'],index=False)
```
